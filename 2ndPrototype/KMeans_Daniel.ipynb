```python
df_test = df.iloc[:,3:14]
df_test
```

```python
scaler = StandardScaler()
scaler.fit(df_test)
```

```python
df_scaled = scaler.transform(df_test)
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=df_test.columns)
df_scaled
```

```python
model = KMeans(n_clusters=8, random_state = 1234)
```

```python
model.fit(df_scaled) 
```

```python
model.labels_
```

```python
labels = model.predict(df_scaled)
labels
```
